```python
import json
import os
```

```python
state = ['Fan on Not Moving', 'idle', 'Fan off Moving', 'Fan on Moving']

datalist=[]
with open('Dynamic_Data.json') as f:
    for jsonObj in f:
        dataDict = json.loads(jsonObj)
        datalist.append(dataDict)

```

```python
KP_MetaData=datalist[1]['Kernel Processes']['MetaData']
SAR=datalist[1]['System Activity Report']
KP_MetaData_list=[]

for item, vals in KP_MetaData.items():
    for key,val in vals.items():
        KP_MetaData_list.append(item+'_'+key)
        

SAR_list=[]
for item, vals in SAR.items():
    if type(vals)==dict:
        for key,val in vals.items():
            SAR_list.append(item+'_'+key)
```

```python

KP_MetaData=datalist[1]['Kernel Processes']['MetaData']
SAR=datalist[1]['System Activity Report']
KP_MetaData_list=[]

for item, vals in KP_MetaData.items():
    for key,val in vals.items():
        KP_MetaData_list.append(item+'_'+key)
        

SAR_list=[]
for item, vals in SAR.items():
    if type(vals)==dict:
        for key,val in vals.items():
            SAR_list.append(item+'_'+key)
logfile=open('features_D.log', 'w')
logfile.write('State,'+','.join(KP_MetaData_list)+','+','.join(SAR_list)+ ',Distance,'+'\n')
```

```python
for data in datalist:
    try:
        #State Labels
        deviceState = data["State"]
        state_label = str(state.index(deviceState))
        
        #Kernel Processes MetaData
        KP_MetaData=data['Kernel Processes']['MetaData']
        KP_MetaData_values=[]
        for key in KP_MetaData_list:                
            if '_' in key:
                header,item=key.split('_')                    
                value=KP_MetaData[header][item]                    
                KP_MetaData_values.append(value)
        
        #System Activity Report
        
        SAR=data['System Activity Report']
        SAR_values=[]
        for key in SAR_list:                
            if '_' in key:
                header,item=key.split('_')                    
                value=SAR[header][item]                    
                SAR_values.append(value)
        print(SAR_values)
        
        #Distance
        distance= data["Distance"]
        
        
        logfile.write(state_label+','+','.join(KP_MetaData_values)+','+','.join(SAR_values)+','+distance+','+'\n')
    except Exception as e:
        print('Something Wrong with Record: ')
```

```python
logfile.close()
os.system('copy features_D.log features_D.csv')
    
    
    
```
